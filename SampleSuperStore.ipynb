```python
Category = df['Category'].unique()
Category_count = df['Category'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = Category,
           y = Category_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Category',fontsize=20)
plt.ylabel('Quantity',fontsize=20)
plt.title('Product Sale',fontsize=30)
plt.show()
```

```python
Country = df['Country'].unique()
Country_count = df['Country'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = Country,
           y = Country_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Country',fontsize=20)
plt.ylabel('Quantity',fontsize=20)
plt.title('Product Sale per Country',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
ax=sns.countplot(x='Category', hue='Sub-Category',data=df)
plt.xlabel('Category',fontsize=17)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(x='Sub-Category',data=df)
plt.title('Sale Sub-Category wise',fontsize=20)
plt.show()
```

```python
df[['Category','Profit']].groupby(["Category"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Profit Category Wise",fontsize=20)
plt.ylabel('Profit',fontsize=15)
plt.xlabel('Category',fontsize=15)
plt.show()
```

```python
df[['Region','Profit']].groupby(["Region"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['red'])
plt.xticks(rotation=50)
plt.title("Profit Region Wise",fontsize=20)
plt.ylabel('Profit',fontsize=15)
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
df[['Sub-Category','Profit']].groupby(["Sub-Category"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['Orange'])
plt.xticks(rotation=50)
plt.title("Profit Sub-Category Wise",fontsize=20)
plt.ylabel('Profit',fontsize=15)
plt.xlabel('Category',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
df['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
data=df[['Category','Sub-Category','Profit']]
data=data.groupby(['Sub-Category','Category'],axis=0).sum().sort_values('Profit',ascending=False).reset_index().head(20)
data
```
